## Auto Grad

```python
import torch as tp

x=tp.tensor(3.0,requires_grad=True)
y=x**2
```

```python
y
```

```python
y.backward()
```

```python
x.grad
```

```python
import torch.nn.functional as f

x=tp.tensor(3.0,requires_grad=True)
y=x**2
z=tp.sin(y)

print(f'x is {x} , y is {y} , z is {z}')
z.backward()
x.grad

x=tp.tensor(6.7)
y=tp.tensor(0.0)
w=tp.tensor(1.0,requires_grad=True)
b=tp.tensor(0.0,requires_grad=True)

z=w*x+b
print(f'z is : {z}')

y_pred=tp.sigmoid(z)
print(y_pred)
loss=f.binary_cross_entropy(y_pred,y)
print('loss',loss)
loss.backward()
print(w.grad)
print(b.grad)
```
